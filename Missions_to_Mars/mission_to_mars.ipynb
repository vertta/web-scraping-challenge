```python
#!pip install splinter[selenium3]  uncomment this line if you do not have selenum3 installed 
# Import splinter, Pandas and beautiful soup
import pandas as pd
from splinter import Browser
from bs4 import BeautifulSoup as soup
from webdriver_manager.chrome import ChromeDriverManager
```

```python
#setup splinter 
executable_path = {'executable_path': ChromeDriverManager().install()}
browser=Browser('chrome', **executable_path, headless = False)
```

```python
#define url for website and then wait for 2 seconds to give the page time to load 
url='https://redplanetscience.com/'
browser.visit(url)
browser.is_element_present_by_css('div.list_text', wait_time=2)

                                  
```

```python
# convert the HTML information in the broswer to a soup object and print parsed information to verify information 
html=browser.html
news_soup = soup(html, 'html.parser')
print(news_soup.prettify())
```

```python
#display the title content by find the div tag and assign the value to variable 'news_title'
slide_elem = news_soup.select_one ('div.list_text')
news_title=slide_elem.find('div', class_ = 'content_title').get_text()
news_title
```

```python

news_paragraph=slide_elem.find('div', class_ = 'article_teaser_body').get_text()
news_paragraph
```

# Hemispheres

```python
#visit the url for hemisphere information
url='https://marshemispheres.com/'
browser.visit(url)
```

```python
#create a list to hold the images and titles 
hemisphere_image_urls = []

# Get a list of all the hemispheres
links = browser.find_by_css('a.product-item img')

#loop through these links, click the link, find the same anchor, return the href 
for i in range(len(links)):
#initialize the array to hold hemisphere information 

    hemisphereInfo = {}
# find the elements on each loop to avad a stale element exception error

    browser.find_by_css('a.product-item img')[i].click()
    
# Find the same image anchor tag and extract the href information
    sample = browser.links.find_by_text('Sample').first
    hemisphereInfo["img_url"] = sample ['href'] 
    
#Get Hemisphere title
    hemisphereInfo['title']  = browser.find_by_css('h2.title').text
#Append hemisphere object to list 
    hemisphere_image_urls.append(hemisphereInfo)

    # Navigate backward
    browser.back()

```
